```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
import csv


# File to Load 
Data = "Data/cost-of-living.csv"
view_data=pd.read_csv(Data)
view_data.head()
```

```python
columns_to_keep =["City","State","Hourly mean wage", "Annual mean wage","Price per Square Meter to Buy Apartment in City Centre","Price per Square Meter to Buy Apartment Outside of Centre","Average Monthly Net Salary (After Tax)",]
```

```python
view_data.drop(columns=view_data.columns.difference(columns_to_keep), inplace=True)
```

```python
view_data.head()
```

```python
view_data_clean = view_data.dropna()
view_data_clean
```

```python
# Clean the "Annual mean wage" column by removing dollar signs and commas
view_data_clean['Annual mean wage'] = view_data_clean['Annual mean wage'].astype(str).str.replace('$', '').str.replace(',', '')

# Convert the cleaned values to numeric
view_data_clean['Annual mean wage'] = pd.to_numeric(view_data_clean['Annual mean wage'], errors='coerce')

# Group the data by "State" and calculate the mean annual wage for each state
state_mean_wage = view_data_clean.groupby("State")["Annual mean wage"].mean().reset_index()

# Sort the states by mean annual wage in descending order
state_mean_wage_sorted = state_mean_wage.sort_values(by="Annual mean wage", ascending=False)

# Select the top 5 states
top_5_states = state_mean_wage_sorted.head(5)

# Create a custom formatting function to add the dollar sign and format values
def format_mean_wage(x, pos):
    return f"${x:,.2f}"

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_5_states["State"], top_5_states["Annual mean wage"])
plt.xlabel("State")
plt.ylabel("Mean Annual Wage")
plt.title ("Top 5 States by Mean Annual Wage")
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Apply the custom formatting function to the y-axis labels
formatter = FuncFormatter(format_mean_wage)
plt.gca().yaxis.set_major_formatter(formatter)

# Show the bar graph
plt.show()

# Save the graph to the data folder
output_file = "Data/top_5_states_mean_wage.png"
plt.savefig(output_file)






```

```python
# Function to format the y-axis labels with a dollar sign
def dollar_formatter(x, pos):
    return f"${x:.2f}"

# Group the data by "State" and calculate the mean annual wage for each state
state_mean_wage = view_data_clean.groupby("State")["Annual mean wage"].mean().reset_index()

# Sort the states by mean annual wage in descending order for top 5 states
top_5_states = state_mean_wage.sort_values(by="Annual mean wage", ascending=False).head(5)

# Sort the states by mean annual wage in ascending order for bottom 5 states
bottom_5_states = state_mean_wage.sort_values(by="Annual mean wage").head(5)

# Create a custom formatting function to add the dollar sign and format values
def format_mean_wage(x, pos):
    return f"${x:,.2f}"

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Create the bars for the top 5 states on the first subplot
top_bars = ax1.bar(top_5_states["State"], top_5_states["Annual mean wage"], color='blue', label='Top 5 States')
ax1.set_xlabel("State")
ax1.set_ylabel("Mean Annual Wage")
ax1.set_title("Top 5 States by Mean Annual Wage")
ax1.tick_params(axis='x', rotation=45)  # Rotate the x-axis labels for better readability

# Apply the custom formatting function to the y-axis labels in the first subplot
ax1.yaxis.set_major_formatter(FuncFormatter(format_mean_wage))

# Create the bars for the bottom 5 states on the second subplot
bottom_bars = ax2.bar(bottom_5_states["State"], bottom_5_states["Annual mean wage"], color='red', label='Bottom 5 States')
ax2.set_xlabel("State")
ax2.set_ylabel("Mean Annual Wage")
ax2.set_title("Bottom 5 States by Mean Annual Wage")
ax2.tick_params(axis='x', rotation=45)  # Rotate the x-axis labels for better readability

# Apply the custom formatting function to the y-axis labels in the second subplot
ax2.yaxis.set_major_formatter(FuncFormatter(format_mean_wage))

# Add a title for the entire figure
fig.suptitle("Comparison of Top 5 and Bottom 5 States by Mean Annual Wage")

# Add a legend to distinguish top and bottom bars
ax2.legend()

# Adjust the layout to avoid overlapping labels
plt.tight_layout()

# Save the entire figure to a file
output_file = "Data/top_bottom_5_states_mean_wage_comparison.png"
plt.savefig(output_file)

# Show the entire figure
plt.show()
```
